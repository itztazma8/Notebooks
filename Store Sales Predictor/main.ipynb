```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Stores.csv')
df.columns
```

```python
# Feature Engineering

def feature1(row):
    if row["Store_Area"]>=700 and row["Store_Area"]<=1300:
        return "Small"
    elif row["Store_Area"]>1300 and row["Store_Area"]<=1900:
        return "Medium"
    else: 
        return "Large"

df["Category"]=df.apply(feature1, axis=1)
```

```python
#Feature Engineering

df["Ratio"]=(df["Daily_Customer_Count"]/df["Store_Area"])
df["Days of Stock"]=(df["Items_Available"]/df["Store_Sales"])*30

df1=df[df["Category"]=="Small"]
df2=df[df["Category"]=="Medium"]
df3=df[df["Category"]=="Large"]

vals=[]
vals.append(int(df1["Store_Sales"].max()))
vals.append(int(df2["Store_Sales"].max()))
vals.append(int(df3["Store_Sales"].max()))


def feature4(row):
    vals=[101780, 105150, 116320]

    if row["Category"]=="Small": 
        return (row["Store_Sales"]/vals[0])
    elif row["Category"]=="Medium":
        return (row["Store_Sales"]/vals[1])
    else:
        return (row["Store_Sales"]/vals[2])

df["Sales Index"]=df.apply(feature4, axis=1)

df.columns
```

```python
#Train-Test setup

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X=df[["Store_Area", "Items_Available", "Daily_Customer_Count", "Ratio", "Days of Stock"
      ,"Sales Index"]]
y=df["Store_Sales"]

X_train, X_test, Y_train, Y_test=train_test_split(X, y, test_size=0.2)

```

```python
#Linear Regression Model

model=LinearRegression()
model.fit(X_train, Y_train)
prediction=model.predict(X_test)
```

```python
#Metrics

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
acc=r2_score(Y_test, prediction)
acc1=mean_absolute_error(Y_test, prediction)
acc2=mean_squared_error(Y_test, prediction)
print("R2: ", acc)
print("MAE: ", acc1)
print("MSE: ", acc2)
```
